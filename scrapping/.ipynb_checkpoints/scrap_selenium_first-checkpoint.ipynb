```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import numpy as np
import pandas as pd
import time
```

```python
path = '/home/penalva/Downloads/chromedriver'
driver = webdriver.Chrome(executable_path=path)
page = driver.get('https://www.linkedin.com/')
user=driver.find_element(By.ID,'session_key')
user.send_keys('dkajah@gmail.com')
passw = driver.find_element(By.ID,'session_password')
passw.send_keys('Aryatara')
sign_in = driver.find_element(By.CLASS_NAME,'sign-in-form__submit-button')
sign_in.click()
driver.get('https://www.linkedin.com/jobs/search/?keywords=%20data%20science%20jobs&location=Brazil')
time.sleep(1)
```

```python
#logged option
descript = []
enterprise = []
where = []
job_type = []
age = []
apply = []

driver.maximize_window()
descript.append(driver.find_element(By.CLASS_NAME,'jobs-description__container').text)
enterprise.append( driver.find_element(By.XPATH,'//div[@class="jobs-unified-top-card__primary-description"]//a').text)
where.append( driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__bullet').text)
try:
    job_type.append( driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__workplace-type').text)
except:
    job_type.append('unknown')
age.append( driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__posted-date').text)

apply_button = driver.find_element(By.CLASS_NAME,'jobs-apply-button')
time.sleep(1)
apply_button.click()

p = driver.current_window_handle
chwd = driver.window_handles
for w in chwd:
    if(w!=p):
        driver.switch_to.window(w)
apply.append(driver.current_url)
driver.close()
chwd = driver.window_handles[0]
driver.switch_to.window(chwd)

for num in np.arange(1,25): 

    job=driver.find_element(By.CLASS_NAME,'jobs-search-two-pane__job-card-container--viewport-tracking-{}'.format(str(num)))

    job.click()

    descript.append(driver.find_element(By.CLASS_NAME,'jobs-description__container').text)
    enterprise.append( driver.find_element(By.XPATH,'//div[@class="jobs-unified-top-card__primary-description"]//a').text)
    where.append( driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__bullet').text)
    try:
        job_type.append( driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__workplace-type').text)
    except:
        job_type.append('unknown')
    age.append( driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__posted-date').text)           
  
    time.sleep(3)
    apply_button = driver.find_element(By.CLASS_NAME,'jobs-apply-button--top-card')
    apply_button.click()                               
        
    p = driver.current_window_handle
    chwd = driver.window_handles
    for w in chwd:
        if(w!=p):
            driver.switch_to.window(w)
    apply.append(driver.current_url)
    driver.close()
    chwd = driver.window_handles[0]
    driver.switch_to.window(chwd)
```

```python
lst = list(zip(descript,enterprise,where,job_type,apply))
df = pd.DataFrame(lst)
df.columns=['Job Description','Enterprise','Where','Job Type','Apply Link']
```
